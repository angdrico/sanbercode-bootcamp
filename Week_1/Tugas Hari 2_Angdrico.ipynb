# Jawaban


### 1.1 List
List merupakan sebuah jenis koleksi data yang dibatasi dengan **kurung siku/bracket "[]"**, **dapat diubah(mutable)**. Dapat menyimpan data yang **duplikat**, dan memiliki **index** serta **berurut** sehingga dapat diakses berdasarkan urutan dan indexnya.

### 1.2 Tuple
Tuple merupakan sebuah jenis koleksi data yang dibatasi dengan **tanda kurung/parentheses "()"** namun **tidak dapat diubah(immutable)**. Dapat menyimpan data **duplikat** dan memiliki **index** serta **berurut** sehingga dapat diakses pula menurut urutan dan indexnya.

### 1.3 Set
Set merupakan sebuah jenis koleksi data yang dibatasi dengan **tanda kurung kurawal/curly bracket "{}"** namun **tidak dapat diubah(immutable)**. **Tidak** dapat menyimpan data **duplikat** , **tidak berurut** dan **tidak memiliki index**.

### 1.4 Dictionary
Dictionary merupakan sebuah jenis koleksi data yang dibatasi dengan **tanda kurung kurawal/curly bracket "{}"** juga, **dapat diubah(mutable)** namun **Tidak** dapat menyimpan data **duplikat**. Memiliki key dan value dalam pengkoleksian datanya. Key bekerja seperti index dalam pemanggilan value yang sudah di-assign terlebih dahulu pada key tersebut

## Soal 6: List Comprehension

Gunakan metode **list comprehension** untuk mencari anggota dari S yang habis di bagi 2, kemudian assign hasilnya dalam bentuk list ke dalam variabel T.

```python
S = [0, 1, 4, 9, 16, 25, 36, 49, 64, 81]

T = [i for i in S if i%2 == 0]

print(T)
```

Expected Output:

[0, 4, 16, 36, 64]

<h2>Soal 11: Loop Dictionary</h2>
    
Lengkapi kode untuk menghasilkan suatu output yang di harapkan

```python
country = { 
           'spain': { 'capital':'madrid', 'population':46.77 },
           'france': { 'capital':'paris', 'population':66.03 },
           'germany': { 'capital':'berlin', 'population':80.62 },
           'norway': { 'capital':'oslo', 'population':5.084 },
           'indonesia' : {'capital':'jakarta', 'population':250}
         }

for a, b in country.items():
    print('Ibukota {} adalah {}'.format(a,country[a]['capital']))
```

Expected Output:
    
Ibukota spain adalah madrid

Ibukota france adalah paris

Ibukota germany adalah berlin

Ibukota norway adalah oslo

Ibukota indonesia adalah jakarta

## Soal 13: Mengubah dan menghapus anggota set

Ubahlah dan hapus anggota set sehingga mendapatkan hasil yang diinginkan


```python
set_1 = {1, 2, 3, 4, 5}

#tambahkan anggota set dengan nilai {6,7,8}
set_1.update([6,7,8])
#print(set_1) #check
set_1.discard(8)
print(set_1)
```
